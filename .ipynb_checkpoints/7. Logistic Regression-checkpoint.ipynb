# Logistic Regression


- Linear Regression만큼 직관적으로 관계성을 모델링 할수 있어보이진 않으나 어느정도의 패턴이 보인다.
- 여기에다가 Linear Regression모델을 적용시키면 어떻게 될까?

- 0과 1을 확률로 생각해보자.
- 0과 1 사이의 중간확률이 없는 이유는 빨간점들은 우리가 이미 Observe한 데이터이기때문에 확률적으로도 100% 아님 0%일 뿐이다.
- 이와 같이 해석해보면 중간에 그려지는 Line이 어느정도 Makes sense하다는걸 알 수 있다.
- 하지만 0,1 바깥의 라인에 대해선 makes sense 하지 않죠? 그래서 모델의 라인을 adjust해줘야 하겠죠


- Logistic Regression을 이와 같이 나타낼 수 있습니다.
- Logistic Regression의 목적도 Linear Regression과 같이 우리의 데이터셋에 가장 잘 Fit 하는 모델라인을 찾는것입니다.

- 위 그림과 같이 Logistic Regression을 이용한 모델이 나오면 Linear Regression과는 다르게 예측값을 확실히 얘기할수가 없습니다.확률일 뿐이죠.

- 그렇기 때문에 이와 같이 중간에 선을 그어 그걸 기준으로 p값에 대한 yes/no 예측을 하게 됩니다.

# Implementation

```python
#Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Import Dataset
dataset = pd.read_csv("Machine Learning A-Z Template Folder/Part 3 - Classification/Section 14 - Logistic Regression/Social_Network_Ads.csv")
```

```python
dataset
```

```python
#살껀지 안살껀지를 예측하는 모델을 찾는것이 우리의 목표입니다.
X = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#Logistic Regression을 import 합시다.
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Evaluation
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred) #par_1: true val ; par_2: pred val
cm
#[[true_yes, false_yes],[false_no,true_no]]
```

```python
#Visualize the result
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

- Red Dots : Users who didnt buy
- Green Dots : Users who bought
- Regions : Prediction Regions predicted by our classifiers

```python
#이제 test set의 예측결과를 시각화 해봅시다.
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
