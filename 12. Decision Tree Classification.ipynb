# Decision Tree Classification
- Decision Tree Regression과 동일한 알고리즘, 단지 predict하는 값이 Contiunuous가 아닌 Categorical.


- 이와같이 초록과 빨강색의 Category로 구성된 데이터들이 있고 이에 따라 잘 구분지을수 있는 Split들을 생성해내는것이 Objective.
- Decision Tree Regression에서는 빨강색의 데이터들만 있었는데 이는 각 split 구역별 y값이 다르기 때문이었죠.



- 유의할 점은 이 예제는 feature column이 두개인(X1,X2) 예제라는 점이다. 더 많아지면 더 복잡한 트리의 형태를 띄게된다.
- Split의 기준은 지난시간에도 설명했듯이 Standard Deviation을 최대한 줄일 수 있는 split을 찾아서 나누게 된다.
- 즉, 각 split한 region들의 데이터들이 variance가 적게 만드는 것. (여기서는 green과 red를 가장 적은 분산을 가지게끔 만드는 역할)
- Decision Tree 자체는 상당히 오래된 방법이지만 Random Forest, Gradient Boosting등과 같은 방법들에 쓰이게 되며 재조명 되었다.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv("Machine Learning A-Z Template Folder/Part 3 - Classification/Section 19 - Decision Tree Classification/Social_Network_Ads.csv")
dataset
```

```python
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

##### Entropy의 개념이 뭐였을까요?
##### Feature Scaling을 해야 할까요?

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.5),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.5))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree Classification (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
#실행 전에 어떻게 시각화가 될지 유추해 봅시다!
```
